```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

# First look
print(matches.shape)
print(matches.head())
```

```python
top_winners = matches['winner'].value_counts().head(10)
print(top_winners)

top_winners.plot(kind='bar', color='Orange', figsize=(10,5))
plt.title("Top 10 IPL Match Winners")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_players = matches['player_of_match'].value_counts().head(10)
print(top_players)

top_players.plot(kind='barh', color='blue', figsize=(10,5))
plt.title("Top 10 Player of the Match Winners")
plt.xlabel("Awards")
plt.tight_layout()
plt.show()
```

```python
matches['toss_match_winner'] = matches['toss_winner'] == matches['winner']
toss_impact = matches['toss_match_winner'].value_counts()
print(toss_impact)

toss_impact.plot(kind='pie', labels=['Toss Won & Won Match', 'Toss Won & Lost Match'],
                 autopct='%1.1f%%', figsize=(6,6))
plt.title("Does Winning Toss Help Win Match?")
plt.tight_layout()
plt.show()
```

```python
top_scorers = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print(top_scorers)

top_scorers.plot(kind='bar', color='green', figsize=(10,5))
plt.title("Top 10 IPL Run Scorers")
plt.xlabel("Player")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
